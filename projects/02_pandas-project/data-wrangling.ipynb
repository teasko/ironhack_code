## the case_no and date  column

```python
# the case_no column contains dates for many of the rows in a yyyy-mm-dd + x form. I filter for those and create 
# a new column "proper dates". this way one can analyze attaks wrt dates whenever they exist

regex_yyyy_mm_dd = r'(19[0-9][0-9]|20[0-9][0-9])[- /.](0[1-9]|1[012])[- /.](0[1-9]|[12][0-9]|3[01])'
regex_yyyy_mm_dd_add = r'(19|20)\d\d[- /.](0[1-9]|1[012])[- /.](0[1-9]|[12][0-9]|3[01])(.+)'

#sharks[sharks.case_number.str.contains(regex_yyyy_mm_dd_add,regex=True)]
cols_with_yyyy_mm_dd = sharks.case_number.str.extract(pat = regex_yyyy_mm_dd)
cols_with_yyyy_mm_dd_true = ~pd.isna(cols_with_yyyy_mm_dd).any(axis=1)

sharks["proper_date"] = np.nan

cols_with_yyyy_mm_dd.columns = ["year","month","day"]
dates = (cols_with_yyyy_mm_dd.loc[cols_with_yyyy_mm_dd_true,"year"]
                                     + " "
                                     + cols_with_yyyy_mm_dd.loc[cols_with_yyyy_mm_dd_true,"month"]
                                     + " "
                                     + cols_with_yyyy_mm_dd.loc[cols_with_yyyy_mm_dd_true,"day"])

dates = pd.to_datetime(dates)
sharks.proper_date = pd.to_datetime(sharks.proper_date)

```

```python
# introducing a new identifyer
sharks["ID"] = sharks.index +1
```

```python
# reorder columns
sharks.loc[cols_with_yyyy_mm_dd_true,["proper_date"]] = dates 
sharks = sharks[["ID","date","proper_date"] + list(sharks.columns)[2:-2]]
```
